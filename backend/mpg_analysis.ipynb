```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/Users/leonleidner/app-coding-agent/backend/uploads/mpg.csv')

# Data Exploration
print("First 5 rows:")
print(data.head())
print("\nData Info:")
print(data.info())
print("\nDescriptive Statistics:")
print(data.describe())

# Data Cleaning
# Handle missing values (if any)
data = data.dropna()

# Visualizations
# Scatterplot of mpg vs weight
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weight', y='mpg', data=data)
plt.title('MPG vs Weight')
plt.xlabel('Weight')
plt.ylabel('Miles Per Gallon')
plt.show()

# Histogram of mpg
plt.figure(figsize=(10, 6))
sns.histplot(data['mpg'], kde=True)
plt.title('Distribution of MPG')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Frequency')
plt.show()

# Boxplot of mpg by origin
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='mpg', data=data)
plt.title('MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('Miles Per Gallon')
plt.show()

# Correlation Analysis
correlation = data[['mpg', 'weight']].corr()
print("\nCorrelation between MPG and Weight:")
print(correlation)

# Statistical Analysis
from scipy import stats
pearson_coef, p_value = stats.pearsonr(data['mpg'], data['weight'])
print("\nPearson Correlation Coefficient:", pearson_coef)
print("P-value:", p_value)

# Recommendations
print("\nRecommendations:")
if abs(pearson_coef) > 0.7:
    print("There is a strong correlation between vehicle weight and fuel efficiency.")
    print("Recommendation: Consider designing lighter vehicles to improve fuel efficiency.")
else:
    print("The correlation between weight and fuel efficiency is not strong.")
    print("Recommendation: Investigate other factors that may influence fuel efficiency.")
```
